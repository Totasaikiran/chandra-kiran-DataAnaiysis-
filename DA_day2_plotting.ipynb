```python
#runs scored by 10 new players [100,50,91,78,89,25,34,19,9,10] wickets taken by name 10 players [1,0,2,0,3,7,8,9,7,5] forrmclusterimport
import numpy as np
runs=np.array([100,50,91,78,89,25,34,19,9,10])
wickets=np.array([1,0,2,0,3,7,8,9,7,5])
plt.scatter(runs,wickets,color="red",marker='^')  #r=red,g=green,b=blue,c=cyan,m=magenta,y=yellow,k=black,w=white
plt.show()                            #'o:circle(default)
#'s':square '^':Triaangle (pointing up) 'v':Triangle (pointing down) '*':star-shaped '+':plus sign 'x': X-shaped '|':vertices
#'':point
```

```python
#plot the score of 2 student in 5 subjects as x axis and mark y axis
S=np.array(['maths','science','english','telugu','hindi','social'])
M1=np.array([75,60,90,65,84,57])
M2=np.array([73,56,89,45,56,67])
plt.plot(S,M1,color="orange",label='jaya kanth')
plt.plot(S,M2,color="blue",label='vamsi')
plt.title("marks and subjects")
plt.legend()
plt.show()
```

```python
S=np.array(['maths','science','english','telugu','hindi','social'])
M1=np.array([75,60,90,65,84,57])
M2=np.array([73,56,89,45,56,67])
plt.subplot(2,1,1)
plt.plot(S,M1,color="green",label='student1')
plt.legend()
plt.subplot(2,1,2)
plt.plot(S,M2,color="purple",label='student2')
plt.legend()
plt.show()
```

```python
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
print("Revenues of company A",A)
print("Revenues of company B",B)
profitA=np.diff(A)
print("profit of A",profitA)
profitB=np.diff(B)
print("profit of B",profitB)
C=np.array(["19-20","20-21","21-22","22-23"])
plt.subplot(1,2,1)
plt.bar(C,profitA,color="black",label="profit A")
plt.legend(loc="best")
plt.subplot(1,2,2)
plt.bar(C,profitB,color="pink",label="profit B")
plt.legend(loc="best")

```
